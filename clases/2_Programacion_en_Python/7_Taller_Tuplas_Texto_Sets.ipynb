### Ejercicio 25: Pipeline de Limpieza (List -> Set -> List)
**Contexto:** Tienes una lista de precios con duplicados y desordenada. Quieres una lista limpia y ordenada.
**Tarea:**
1. Convierte la lista `precios_raw` a set para quitar duplicados.
2. Convierte ese set de nuevo a lista con list().
3. Ordena la lista resultante usando sorted().
Guarda el resultado final en `precios_limpios`.

```python
precios_raw = [100, 200, 100, 50, 200]
```

```python
### TU CÓDIGO AQUÍ ###
precios_raw = set(precios_raw)
precios_raw = list(precios_raw)
precios_limpios = sorted(precios_raw)
print(precios_limpios)
```
